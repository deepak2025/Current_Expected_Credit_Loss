```python
#Predictions of chargeoff of Ameris Bank using factors for which it has higher correlation
```

```python
X=chargeoff_national[['Unemp_Rate','SOUTH STATE BANK (540926)','HOMETRUST BANK (212577)','LIVE OAK BANKING COMPANY (3650808)']]
Y=chargeoff_national['AMERIS BANK (764030)']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.5)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
ameris=LinearRegression()
```

```python
ameris.fit(X_train,Y_train)
```

```python
print(ameris.intercept_)
```

```python
ameris.coef_
```

```python
predictions=ameris.predict(X_test)
```

```python
print(predictions.mean(),predictions.std())
```

```python
plt.scatter(Y_test,predictions)
```

```python
sns.distplot((Y_test-predictions))
```

```python
from sklearn import metrics
```

```python
metrics.r2_score(Y_test,predictions)
```
